```python
#### Amdhal Law based on Benny course
"""
T(p) = Ts + Tp/p

- T(p) est le temps total d'execution sur un système avec p processeurs.
- Ts est le temps pris par la partie séquentiel du code (Ici ce serait le temps de une simu. 
Mais ce temps n'est le même selon le nb de processus parallèle.)
- Tp est le temps pris par la partie parallèle du code pour s'executer sur un processeur
(Dans notre cas, il n'y a rien ?)
- p est le nombre de processeurs.

"""
#### Speedup :
"""

Sp = T1 / Tp

T1 la latence pour un coeur de calcul (Le temps pris pour terminer la tache sur 1 seul coeur)
Tp la latence pour p coeurs de calcul (Le temps pris pour terminer la tache sur p coeurs)

La tache doit etre identique. (64 simu covid par exemple)

"""
#### Efficience :
"""
Efficiency = Sp/P = T1/ P*Tp

La valeur doit tendre vers 1. Mais en réalité elle s'en éloigne.

"""
# La formule que je vais utiliser est le speedup

```
